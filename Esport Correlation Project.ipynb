## 1. Importing libraries and data <a name="introduction"></a>

```python
#We begin by importing our packages and data needed 

import pandas as pd 
import seaborn as sns
import numpy as np 
import matplotlib 
import matplotlib.pyplot as plt 

#Reading our CSV's from file path 
df1 = pd.read_csv(r'/Users/mattyro/Desktop/GeneralEsportData.csv')

df2 = pd.read_csv(r'/Users/mattyro/Desktop/Twitch_game_data2021.csv')

#Let's take a look at the data and some basic information for df1 using python pandas.

print(df1.shape) #What is the data row count?


print(df1.describe()) #What is the distribution of data?

print(df1.info()) #What data types are in our data?

#Bring up the dataframe 
df1




```

```python
#Similarly, we take a look at df2. 

print(df2.shape) 


print(df2.describe()) 

print(df2.info()) 

#Bring up the dataframe 
df2
```

```python
#Percentage of missing data 
for i in df1.columns:
    pct_missing = np.mean(df1[i].isnull())
    print('{} - {}%'.format(i,pct_missing*100))

for j in df2.columns:
    pct_missing = np.mean(df2[j].isnull())
    print('{} - {}%'.format(j,pct_missing*100))
```

## 2. Data cleaning <a name="paragraph1"></a>

```python
#Let's drop some columns which are not necessary
df1 = df1.drop(['ReleaseDate','OnlineEarnings'], axis = 1)
df2 = df2.drop(['Hours_streamed','Streamers','Avg_viewers'], axis = 1)

#Bring up the first 5 rows in each dataframe to see the changes 
print(df1.head())

print(df2.head())
```

```python
#Merging the dataframes 
#In order to make things more simpler for when we want to show correlations, let's merge the two dataframes into one 
df3 =pd.merge(df1, df2, on ='Game')
df3 
```

```python
#Grouping the data using Pandas
#Merging the data gives us unwanted duplicates in response to the month column, so let's group the data in a more compact format 
df3 = df3.groupby(['Game','Genre','TotalEarnings','TotalPlayers','TotalTournaments'], as_index = False).agg(TotalHoursWatched = ('Hours_watched','sum'))

df3
```

```python
#Sorting the data 
#Since esports is a monetary industry, we are interested in seeing which "successful" games are lucrative in esports
df3.sort_values(by = ['TotalEarnings'], inplace = True, ascending = False) #sort by total earnings 

df3
```

```python
#Remove zero earning titles 
df3 = df3[df3['TotalEarnings']>0]

df3
```

## 3. Data Visualization <a name="paragraph2"></a> 

```python
#Heatmap using seaborn 
#Let's take a look at a correlation matrix to show which numerical variables are strongly correlated 
corrmatrix = df3.corr(method = 'pearson')
sns.heatmap(corrmatrix, annot= True)
plt.title('Correlation Matrix')
plt.show()

```

```python
#Scatterplot showing relationship between total tournaments and total earnings
plt.scatter(x=df3['TotalTournaments'], y=df3['TotalEarnings'], color= 'red')

plt.title('Total Tournaments vs Total Earnings')

plt.xlabel('Number of Tournaments')

plt.ylabel('Total Tournament Earnings')

plt.show()
```

```python
#Suppose we want to see the general trend of our scatterplot, we can see that it is linearly increasing.

sns.regplot(x='TotalTournaments', y= 'TotalEarnings', data=df3, scatter_kws={"color": "red"}, line_kws={"color": "blue"}).set(title='Total Tournaments vs Total Earnings')
```

```python
#Bar graph of how total earnings each genre produces 
earnings_per_genre = df3.groupby(['Genre'])['TotalEarnings'].sum()

plt.title("Total Earnings in each Category")
plt.xlabel('Genre')
plt.ylabel('Earnings')

earnings_per_genre.plot(kind='bar', color= 'teal')
plt.show()
```
